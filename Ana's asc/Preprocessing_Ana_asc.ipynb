```python
import os
import re
import itertools
import numpy as np
from stat import S_ISREG, ST_CTIME, ST_MODE,ST_MTIME
import os, sys, time
from shutil import copyfile
import shutil

fpath = 'Promoter collection 0.4% glc vs 0.2% glc'
lfs = os.listdir(fpath)
plates = []
subfolders=[]
for s in lfs:
    plate = map(int,re.findall(r'\d+', s))
    print plate
    if len(plate)>0:
        subfolders.append( os.path.join(fpath,s))
        plates.append(plate)


flatten = itertools.chain.from_iterable
flat_l = list(flatten(plates))
n_l = np.asarray(flat_l)
plate_max = n_l.max()
half = plate_max/2

output = ['rep1','rep2','rep3']

for rep in output:
    directory = os.path.join(rep,'asc')
    if os.path.exists(directory):
        shutil.rmtree(directory)
    os.mkdir(directory)




for i, sub in enumerate(subfolders):
    name = plates[i]
    print sub, i, name
    
    entries = []
    for root, dirs, files in os.walk(sub):
        for fn in files:            
            entries.append(os.path.join(root,fn))
    
    # get all entries in the directory w/ stats
    entries = ((os.stat(path), path) for path in entries)

    # leave only regular files, insert creation date
    entries = ((stat[ST_MTIME], path)
               for stat, path in entries if S_ISREG(stat[ST_MODE]))
    #NOTE: on Windows `ST_CTIME` is a creation date 
    #  but on Unix it could be something else
    #NOTE: use `ST_MTIME` to sort by a modification date

    sorted_mt_path = []
    for cdate, path in sorted(entries):
        sorted_mt_path.append(path)
        #print path
        #print time.ctime(cdate), os.path.basename(path)
        
    for i, path in enumerate(sorted_mt_path):
        base_name = os.path.basename(path)
        base_name = base_name.split('.')[0]
        if i%12 == 0:
            plate_name = name[0]
            start = i
        elif i%6 == 0:
            plate_name = name[1]
            start = i
        
        if (i-start)%6 == 0:
            dst = os.path.join("rep1",'asc',base_name+"_Plate"+str(plate_name)+'.asc')
        elif (i-start)%6 == 2:
            dst = os.path.join("rep2",'asc',base_name+"_Plate"+str(plate_name)+'.asc')
        elif (i-start)%6 == 4:
            dst = os.path.join("rep3",'asc',base_name+"_Plate"+str(plate_name)+'.asc')
        elif (i-start)%6 == 1:
            dst = os.path.join("rep1",'asc',base_name+"_Plate"+str(plate_name+plate_max)+'.asc')
        elif (i-start)%6 == 3:
            dst = os.path.join("rep2",'asc',base_name+"_Plate"+str(plate_name+plate_max)+'.asc')
        elif (i-start)%6 == 5:
            dst = os.path.join("rep3",'asc',base_name+"_Plate"+str(plate_name+plate_max)+'.asc')
        
        copyfile(path, dst)
    
```
